```python
"""
 1. Exercício
 Criação
 
 - Crie uma classe chamada Aluno com os seguintes atributos: nome e nota.

 - Adicione um construtor (init) que inicialize os atributos nome e nota com dois parâmetros (nome e nota). Inclua um print dentro 
 do construtor para exibir a mensagem: "O aluno foi criado com sucesso".

- Adicione o método (str) para retornar uma string com os valores dos atributos nome e nota no formato: "Aluno: [nome], Nota: [nota]".

- Crie um método chamado qualificação que imprima se o aluno passou ou reprovou com base na sua nota.


- Implemente a técnica do encapsulamento, com a implementação dos atributos para 
privado: __nome e __nota, e criação dos métodos getters e setters



"""
```

A diferença entre getters e setters está na finalidade de cada um:

Getters: São usados para obter o valor de um atributo privado. Eles retornam o valor do atributo para que possa ser usado em outras partes do código. Por isso, eles precisam de um return para devolver o valor.

Setters: São usados para definir ou modificar o valor de um atributo privado. Eles recebem um valor como parâmetro e o atribuem ao atributo privado. Não há necessidade de retornar nada, pois a função do setter é apenas atualizar o valor do atributo.




```python
class Aluno:

    nome = ""  # abritutos
    nota = 0
    
    def __init__(self, new_nome, new_nota ):
        
        self.__nome = new_nome
        self.__nota = new_nota

        print(f"O aluno foi criado com sucesso")

    # métodos getters

    def get_nome(self):
        return self.__nome
    
    def get_nota(self):
        return self.__nota
    

    ## metodo setters
    def set_nota(self, nota):
        
        self.__nota = nota
        return(f"{self.get_nome()} fez a recuperação e ficou com a nota {nota}")

    def __str__(self):
        return (f"Aluno: {self.get_nome()}, Nota:{self.get_nota()}")
    
    def qualificacao(self):

        try:

            nota = float(self.get_nota())
            if nota < 60:
                return(f"Esse aluno {self.get_nome()} foi reprovado {nota}")
            
            else:
                return(f"Passou! ´{self.get_nome()}´. Parabéns ")

        except ValueError:
            return("Não deu")
```

```python

# 3.1 
class Produto:

    codigo = 0 # aatributos
    nome = ""
    preço = 0
    tipo = ""
    quantidade_iten = 0

    def __init__(self, new_codigo, new_nome, new_preço, new_tipo): #new_quantidade):
        self.__codigo = new_codigo
        self.__nome = new_nome
        self.__preço = new_preço
        self.__tipo = new_tipo
        self.__quantidade_iten = 0
        self.__lista = []

        print(f"Produto cadastrado com sucesso!")

    ## metodo getters e setters
    def get_codigo(self):
        return self.__codigo
    
    def get_nome(self):
        return self.__nome
    
    def get_preço(self):
        return self.__preço 
    
    def get_tipo(self):
        return self.__tipo
    

    ### setter

    def set_codigo(self, codigo):
        self.__codigo = codigo 
        
    def set_nome(self, nome):
        self.__nome = nome
    
    def set_preço(self, preço):
        self.__preço = preço

    def set_tipo(self, tipo):
        self.__tipo = tipo
    
    def __str__(self):
        try:
            return(f"Produto [{self.get_codigo()}]: [{self.get_nome()}], Preço: [{self.get_preço()}], Tipo: [{self.get_tipo()}]")
        except ValueError:
            print("Erro", ValueError)

    def calcula_total(self, quantidade=0):

        try:
            quantidade = int(quantidade)

        except ValueError:
            print(f" Não consegui! Digite apenas numero")
        valorTotal = quantidade * self.get_preço()
        print(f"{quantidade} X {self.get_preço()} = {valorTotal:.2f}. Obrigado pela preferencia")
        return valorTotal

class Pedido(Produto):

    produtos = []
    quantidade = 0

    def __init__(self, produtos, quantidades):
        self.produtos = produtos
        self.quantidades = quantidades

        print(f"Criado objeto com sucesso!")

    def total_pedido(self):
        total = 0

        for produto, quantidade in zip(self.produtos, self.quantidades):
            total += produto.get_preço() * quantidade

        return round(total, 2)
    
    def mostrar_pedido(self):

        #TODO estamos com um bug aqui # encontrei, se fizer o return, ele buga, mostra so 1 item da lista
        for produto, quantidade in zip(self.produtos, self.quantidades):
            print(f"Produto: [{produto.get_nome()}], Quantidade: [{quantidade}]")

    def mostrar_pedido_outra_forma(self):
        pedidos = [f"Produto: [{produto.get_nome()}], Quantidade: [{quantidade}]" for produto, quantidade in zip(self.produtos, self.quantidades)]
        for pedido in pedidos:
            print(pedido)

    def mostrar_pedido_outra_forma_com_map(self):
        def mostrar(produto, quantidade):
            print(f"Produto: [{produto.get_nome()}], Quantidade: [{quantidade}]")
    
        list(map(mostrar, self.produtos, self.quantidades))

    
    def mostrar_pedido_com_for(self):
        for i in range(len(self.produtos)):
            produto = self.produtos[i]
            quantidade = self.quantidades[i]
            print(f"Produto: [{produto.get_nome()}], Quantidade: [{quantidade}]")

    def mostrar_pedido_com_enumerete(self):
        for i in range(len(self.produtos)): produto = self.produtos[i] 
        quantidade = self.quantidades[i] 
        print(f"Produto: [{produto.get_nome()}], Quantidade: [{quantidade}]")
```
