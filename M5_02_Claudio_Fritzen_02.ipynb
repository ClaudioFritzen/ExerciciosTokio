```python
# Load the workbook

resumo = openpyxl.load_workbook('dados.xlsx')
folha = resumo.active

# pegando a quantidade de notas insuficientes e suficientes por materia

qtd_suficiente_matematica = 0
qtd_suficente_portugues = 0
qtd_suficiente_ciencias = 0

qtd_insuficiente_matematica = 0
qtd_insuficiente_portugues = 0
qtd_insuficente_ciencias = 0

for linha in folha.iter_rows(min_row=2, max_row=6, min_col=6, max_col=8):
    for celula in linha:
        if celula.column == 6:
            if celula.value == "Suficiente":
                qtd_suficiente_matematica += 1
            else:
                qtd_insuficiente_matematica += 1
        elif celula.column == 7:
            if celula.value == "Suficiente":
                qtd_suficente_portugues += 1
            else:
                qtd_insuficiente_portugues += 1
        elif celula.column == 8:
            if celula.value == "Suficiente":
                qtd_suficiente_ciencias += 1
            else:
                qtd_insuficente_ciencias += 1

print(f"Matematica: {qtd_suficiente_matematica} alunos suficientes e {qtd_insuficiente_matematica} alunos insuficientes")
print(f"Portugues: {qtd_suficente_portugues} alunos suficientes e {qtd_insuficiente_portugues} alunos insuficientes")
print(f"Ciencias: {qtd_suficiente_ciencias} alunos suficientes e {qtd_insuficente_ciencias} alunos insuficientes")

# vamos criar uma nova planilha com esses dados

nova_planilha = openpyxl.Workbook()
nova_folha = nova_planilha.active

nova_folha['A1'] = 'Matéria'
nova_folha['B1'] = 'Suficiente'
nova_folha['C1'] = 'Insuficiente'

nova_folha['A2'] = 'Matemática'
nova_folha['B2'] = qtd_suficiente_matematica
nova_folha['C2'] = qtd_insuficiente_matematica

nova_folha['A3'] = 'Português'
nova_folha['B3'] = qtd_suficente_portugues
nova_folha['C3'] = qtd_insuficiente_portugues

nova_folha['A4'] = 'Ciências'
nova_folha['B4'] = qtd_suficiente_ciencias
nova_folha['C4'] = qtd_insuficente_ciencias

## pegando a quantidade de alunos 
qtd_alunos = 0 

for linha in folha.iter_rows(min_row=2, max_row=6, min_col=1, max_col=1):
    for celula in linha:
        qtd_alunos += 1
print(f"Quantidade de alunos: {qtd_alunos}")


# abrimos a planilha para salvar

nova_folha["D1"] = f"Total de alunos"
nova_folha["D2"] = f"{qtd_alunos}"
nova_folha["D3"] = f"Total de alunos: {qtd_alunos}"
nova_planilha.save('dados_analise.xlsx')

```

Criação

Crie o ficheiro "dados.csv" com os dados fornecidos, utilizando a biblioteca csv.

Produto     Categoria       Preço       Quantidade

Teclado     Informatica     20.00       100
Mouse       Informatica     15.00       200
Cadeira     Móveis          100.00      50
Caneca      Utensílios      5.00        300
Mesa        Móveis          150.00      20
Cadeira     Móveis          250.00      20

Ordene os dados pela coluna "Preço", do maior para o menor.
Filtrar os dados por produtos com preço acima de 25.
Após ordenar e filtrar, salve os resultados em um novo ficheiro CSV chamado "dados_filtrados.csv".



```python
# abrindo um arquivo csv 

import csv

# criando um arquivo csv
with open('vendas.csv', 'w', newline='') as arquivo:
    escritor = csv.writer(arquivo)
    escritor.writerow(['Produto', 'Categoria', 'Preço', 'Quantidade'])
    escritor.writerow(['Teclado', 'Informatica', 20.00, 100])
    escritor.writerow(['Mouse', 'Informatica', 15.00, 200])   
    escritor.writerow(['Cadeira', 'Móveis', 100.00, 50])
    escritor.writerow(['Caneca', 'Utensílios', 5.00, 300])
    escritor.writerow(['Mesa', 'Móveis', 150.00, 20])
    escritor.writerow(['Móveis', 'Móveis', 250.00, 20])

# salvando o arquivo com latin1
with open('vendas.csv', 'r', encoding='latin1') as arquivo:
    leitor = csv.reader(arquivo)
    for linha in leitor:
        print(linha)  

# queremos fazer um filtro por preço do maior a o menor
with open('vendas.csv', 'r', encoding='latin1') as arquivo:
    leitor = csv.reader(arquivo)
    header = next(leitor)  # skip the header row
    sorted_data = sorted(leitor, key=lambda x: float(x[2]), reverse=True)
    print(f"Lista ordenada por preço do maior ao menor preço")
    print(header)  # print the header row
    for linha in sorted_data:
        print(linha)

# Salvando os dados ordenados em um novo arquivo CSV
with open('filtro_do_maior_ao_menor.csv', 'w', newline='', encoding='latin1') as arquivo_filtrado:
    escritor = csv.writer(arquivo_filtrado)
    escritor.writerow(header)  # write the header row
    escritor.writerows(sorted_data)

# Caminho para o arquivo CSV
caminho_arquivo_csv = "vendas.csv"

# Abrir e ler o arquivo CSV
with open(caminho_arquivo_csv, mode='r', encoding='latin1') as arquivo_csv:
    leitor_csv = csv.reader(arquivo_csv)
    cabecalho = next(leitor_csv)  # Ler o cabeçalho
    filtro = 25.0  # Valor mínimo para o filtro

    # Filtrar as linhas usando list comprehension
    linhas_filtradas = [linha for linha in leitor_csv if float(linha[2].replace(',', '.')) > filtro]

    # Mostrar as linhas filtradas
    print(f"Preço filtrado pelo valor {filtro}")
    # Mostrar o cabeçalho
    print(cabecalho)
    for linha in linhas_filtradas:
        print(linha)

# Salvando os dados filtrados em um novo arquivo CSV
with open('dados_filtrados.csv', 'w', newline='', encoding='latin1') as arquivo_filtrado:
    escritor = csv.writer(arquivo_filtrado)
    escritor.writerow(cabecalho)  # write the header row
    escritor.writerows(linhas_filtradas)

```

Mês     Vendas

Janeiro 2000
Fevereiro   3000
Março   2500
Abril   4000
Maio    3500

Utilize a biblioteca openpyxl para gerar um gráfico de linha que mostre a evolução das vendas ao longo dos meses.
Adicione o gráfico à folha, posionando-o abaixo da tabela de vendas.
Salve o ficheiro como "vendas_com_gráfico.xlsx".

```python
import openpyxl
```

```python
# vamos criar um novo arquivo
wb = openpyxl.Workbook()
wb.create_sheet('vendas.xlsx')
wb.save('vendas.xlsx')

# abrindo o arquivo
vendas = openpyxl.load_workbook('vendas.xlsx')
folha = vendas.active

# adicionando os dados
folha.append(['Mês', 'Vendas'])
folha.append(['Janeiro', 2000])
folha.append(['Fevereiro', 3000])
folha.append(['Março', 2500])
folha.append(['Abril', 4000])
folha.append(['Maio', 3500])
folha.append(['Junho', 4500])

# salvando o arquivo
vendas.save('vendas.xlsx')

# fechando o arquivo    

from openpyxl.chart import BarChart, Reference, Series
folha = vendas.active
# criando um gráfico
chart = BarChart()
chart.type = "col"
chart.style = 10
chart.title = "Vendas por mês"
chart.x_axis.title = "Mês"
chart.y_axis.title = "Vendas"

# criando as referências
data = Reference(folha, min_col=2, min_row=1, max_row=7, max_col=2)
categories = Reference(folha, min_col=1, min_row=2, max_row=7)

# adicionando as referências ao gráfico
chart.add_data(data, titles_from_data=True)
chart.set_categories(categories)

# adicionando o gráfico à planilha
folha.add_chart(chart, "D1")

# salvando o arquivo
vendas.save('vendas.xlsx')


```
