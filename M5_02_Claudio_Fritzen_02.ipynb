```python
import openpyxl
```

```python
# vamos criar um arquivo excel com esse seguinte conteúdo
# | Nome | Idade| Nota Matematica | Nota Português | Nota Ciencia |
# | João | 16   | 8               | 7              | 6           |
# | Maria| 17   | 17              | 19             | 14          |
# | Pedro| 16   | 6               | 9              | 10          |
# | Ana  | 18   | 15              | 20             | 17          |


# | José | 16   | 6               | 5              | 7           |

# criando um arquivo excel
wb = openpyxl.Workbook()
# selecionando a primeira planilha
sheet = wb.active
# nomeando a planilha
sheet.title = 'dados.xlsx'

# criando os cabeçalhos
sheet['A1'] = 'Nome'
sheet['B1'] = 'Idade'
sheet['C1'] = 'Nota Matematica'
sheet['D1'] = 'Nota Português'
sheet['E1'] = 'Nota Ciencia'

# criando os dados
sheet['A2'] = 'João'
sheet['B2'] = 16
sheet['C2'] = 8
sheet['D2'] = 7
sheet['E2'] = 6

sheet['A3'] = 'Maria'
sheet['B3'] = 17
sheet['C3'] = 17
sheet['D3'] = 19
sheet['E3'] = 14

sheet['A4'] = 'Pedro'
sheet['B4'] = 16
sheet['C4'] = 6
sheet['D4'] = 9
sheet['E4'] = 10

sheet['A5'] = 'Ana'
sheet['B5'] = 18
sheet['C5'] = 15
sheet['D5'] = 20
sheet['E5'] = 17

# salvando o arquivo
wb.save('dados.xlsx')
print('Arquivo criado com sucesso')


```

```python
modificando = wb.active.cell(row=1, column=6, value="Avaliação Matematica")
modificando = wb.active.cell(row=1, column=7, value="Avaliação Português")
modificando = wb.active.cell(row=1, column=8, value="Avaliação Ciências")


# faremos o for por cada linha da planilha

for linha in range(2, wb.active.max_row + 1):
    matematica = wb.active.cell(row=linha, column=3).value
    portugues = wb.active.cell(row=linha, column=4).value
    ciencias = wb.active.cell(row=linha, column=5).value

    """ print(type(f'Nota de Matemática: {matematica}'))
    print(f'Nota de Português: {portugues}')
    print(f'Nota de Ciências: {ciencias}') """

    # convertendo os valores para numeros fluantes
    matematica = float(matematica) if matematica is not None else 0
    portugues = float(portugues) if portugues is not None else 0
    ciencias = float(ciencias) if ciencias is not None else 0

    # print com as alterações
    print(f'Nota de Matemática: {matematica}')
    #print(f'Nota de Português: {portugues}')
    #print(f'Nota de Ciências: {ciencias}') """

    # verificar se as notas são maiores que 10, > 10 = "Suficente" < 10 = "Insuficiente"

    avaliacao_matematica = "Suficiente" if matematica >= 10 else "Insuficiente"
    avalicao_portugues = "Suficiente" if portugues >= 10 else "Insuficiente"
    avaliacao_ciencias = "Suficiente" if ciencias >= 10 else "Insuficiente"

    print(f'Avaliação de Matemática: {avaliacao_matematica}')

    # adicionando as avaliações na planilha
    wb.active.cell(row=linha, column=6, value=avaliacao_matematica)
    wb.active.cell(row=linha, column=7, value=avalicao_portugues)
    wb.active.cell(row=linha, column=8, value=avaliacao_ciencias)

wb.save('dados.xlsx')
```

```python
# Load the workbook

resumo = openpyxl.load_workbook('dados.xlsx')
folha = resumo.active

# pegando a quantidade de notas insuficientes e suficientes por materia

qtd_suficiente_matematica = 0
qtd_suficente_portugues = 0
qtd_suficiente_ciencias = 0

qtd_insuficiente_matematica = 0
qtd_insuficiente_portugues = 0
qtd_insuficente_ciencias = 0

for linha in folha.iter_rows(min_row=2, max_row=6, min_col=6, max_col=8):
    for celula in linha:
        if celula.column == 6:
            if celula.value == "Suficiente":
                qtd_suficiente_matematica += 1
            else:
                qtd_insuficiente_matematica += 1
        elif celula.column == 7:
            if celula.value == "Suficiente":
                qtd_suficente_portugues += 1
            else:
                qtd_insuficiente_portugues += 1
        elif celula.column == 8:
            if celula.value == "Suficiente":
                qtd_suficiente_ciencias += 1
            else:
                qtd_insuficente_ciencias += 1

print(f"Matematica: {qtd_suficiente_matematica} alunos suficientes e {qtd_insuficiente_matematica} alunos insuficientes")
print(f"Portugues: {qtd_suficente_portugues} alunos suficientes e {qtd_insuficiente_portugues} alunos insuficientes")
print(f"Ciencias: {qtd_suficiente_ciencias} alunos suficientes e {qtd_insuficente_ciencias} alunos insuficientes")

# vamos criar uma nova planilha com esses dados

nova_planilha = openpyxl.Workbook()
nova_folha = nova_planilha.active

nova_folha['A1'] = 'Matéria'
nova_folha['B1'] = 'Suficiente'
nova_folha['C1'] = 'Insuficiente'

nova_folha['A2'] = 'Matemática'
nova_folha['B2'] = qtd_suficiente_matematica
nova_folha['C2'] = qtd_insuficiente_matematica

nova_folha['A3'] = 'Português'
nova_folha['B3'] = qtd_suficente_portugues
nova_folha['C3'] = qtd_insuficiente_portugues

nova_folha['A4'] = 'Ciências'
nova_folha['B4'] = qtd_suficiente_ciencias
nova_folha['C4'] = qtd_insuficente_ciencias

## pegando a quantidade de alunos 
qtd_alunos = 0 

for linha in folha.iter_rows(min_row=2, max_row=6, min_col=1, max_col=1):
    for celula in linha:
        qtd_alunos += 1
print(f"Quantidade de alunos: {qtd_alunos}")


# abrimos a planilha para salvar

nova_folha["D1"] = f"Total de alunos"
nova_folha["D2"] = f"{qtd_alunos}"
nova_folha["D3"] = f"Total de alunos: {qtd_alunos}"
nova_planilha.save('dados_analise.xlsx')

```
